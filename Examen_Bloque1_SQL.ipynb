## Parte C: Consultas SQL (10 Puntos)

Responda cada pregunta escribiendo la consulta SQL correspondiente en la variable `query`.

### Consulta 1
Liste el nombre y correo de todos los estudiantes registrados.

```python
query_1 = """
SELECT nombre, correo 
FROM estudiantes;
"""

ejecutar_consulta(query_1)
```

### Consulta 2
Obtenga los estudiantes cuya edad sea mayor a 20 años.

```python
query_2 = """
SELECT * FROM estudiantes 
WHERE edad > 20;
"""

ejecutar_consulta(query_2)
```

### Consulta 3
Muestre el nombre de los cursos con una duración mayor a 30 horas.

```python
query_3 = """
SELECT nombre 
FROM cursos 
WHERE duracion_horas > 30;
"""

ejecutar_consulta(query_3)
```

### Consulta 4
Liste los nombres de los estudiantes junto con los cursos en los que están inscritos.

```python
query_4 = """
SELECT e.nombre AS Estudiante, c.nombre AS Curso
FROM inscripciones i
INNER JOIN estudiantes e ON i.id_estudiante = e.id_estudiante
INNER JOIN cursos c ON i.id_curso = c.id_curso;
"""

ejecutar_consulta(query_4)
```

### Consulta 5
Muestre cuántos estudiantes están inscritos en cada curso.

```python
query_5 = """
SELECT c.nombre AS Curso, COUNT(i.id_estudiante) AS total_inscritos
FROM cursos c
LEFT JOIN inscripciones i ON c.id_curso = i.id_curso
GROUP BY c.nombre;
"""

ejecutar_consulta(query_5)
```

### Consulta 6
Obtenga el promedio de edad de los estudiantes.

```python
query_6 = """
SELECT AVG(edad) AS edad_promedio 
FROM estudiantes;
"""

ejecutar_consulta(query_6)
```

### Consulta 7
Liste los cursos ordenados por duración de mayor a menor.

```python
query_7 = """
SELECT nombre, duracion_horas 
FROM cursos 
ORDER BY duracion_horas DESC;
"""

ejecutar_consulta(query_7)
```

### Consulta 8
Muestre los estudiantes inscritos en el curso llamado “Bases de Datos”.

```python
query_8 = """
SELECT e.nombre
FROM estudiantes e
INNER JOIN inscripciones i ON e.id_estudiante = i.id_estudiante
INNER JOIN cursos c ON i.id_curso = c.id_curso
WHERE c.nombre = 'Bases de Datos';
"""

ejecutar_consulta(query_8)
```

### Consulta 9
Obtenga los cursos que tienen al menos un estudiante inscrito.

```python
query_9 = """
SELECT DISTINCT c.nombre
FROM cursos c
INNER JOIN inscripciones i ON c.id_curso = i.id_curso;
"""

ejecutar_consulta(query_9)
```

### Consulta 10
Muestre la cantidad total de inscripciones registradas.

```python
query_10 = """
SELECT COUNT(*) AS total_inscripciones 
FROM inscripciones;
"""

ejecutar_consulta(query_10)
```
