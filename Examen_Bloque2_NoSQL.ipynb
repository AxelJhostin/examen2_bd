## Parte C: Consultas NoSQL (10 Puntos)

Resuelva las siguientes consultas utilizando los métodos de la colección (find, update_one, delete_one, aggregate, etc.).

### Consulta 1
Obtenga todos los estudiantes registrados.

```python
resultado_1 = collection.find({})

for doc in resultado_1:
    print(doc)
```

### Consulta 2
Muestre únicamente los nombres y correos de los estudiantes.

```python
resultado_2 = collection.find({},{"nombre": 1, "correo": 1, "_id": 0})

for doc in resultado_2:
    print(doc)
```

### Consulta 3
Obtenga los estudiantes cuya edad sea mayor a 20 años.

```python
filtro = {"edad": {"$gt": 20}}
resultado_3 = collection.find(filtro)

for doc in resultado_3:
    print(doc)
```

### Consulta 4
Muestre los estudiantes inscritos en el curso “Bases de Datos”.

```python
filtro = {"cursos.nombre":"Bases de Datos"}
resultado_4 = collection.find(filtro)

for doc in resultado_4:
    print(doc)
```

### Consulta 5
Obtenga los estudiantes con calificación mayor o igual a 80 en algún curso.

```python
filtro = {"cursos.calificacion":{"$gte":80}}
resultado_5 = collection.find(filtro)

for doc in resultado_5:
    print(doc)
```

### Consulta 6
Agregue un nuevo curso a un estudiante específico.

```python
collection.update_one({"nombre":"Hernandez Axel"},{"$push":{"cursos":{"nombre":"Cloud Computing","duracion_horas":15,"calificacion":88}}})
resultado_6 = collection.find({"nombre":"Hernandez Axel"},{"_id":0,"nombre":1,"cursos":1})

for doc in resultado_6:
    print(doc)

```

### Consulta 7
Actualice la calificación de un curso para un estudiante.

```python
collection.update_one({"nombre":"Hernandez Axel","cursos.nombre":"Cloud Computing"},{"$set":{"cursos.$.calificacion":100}})
resultado_7 = collection.find({"nombre":"Hernandez Axel"},{"_id":0,"nombre":1,"cursos":1})

for doc in resultado_7:
    print(doc)


```

### Consulta 8
Elimine un estudiante según su correo electrónico.

```python
filtro = {"correo": "adolfo@mail.com"}
collection.delete_one(filtro)
resultado_8 = collection.find({})

for doc in resultado_8:
    print(doc)


```

### Consulta 9
Cuente cuántos estudiantes tienen más de un curso registrado.

```python
resultado_9 = {"cursos.1": {"$exists": True}}
cantidad = collection.count_documents(resultado_9)

print(f"La cantidad de estudiantes con más de un curso es: {cantidad}")
```

### Consulta 10
Obtenga los nombres de los estudiantes inscritos en cualquiera de los cursos: “SQL” o “MongoDB”.

```python
filtro = {"cursos.nombre": {"$in": ["SQL", "MongoDB"]}}
proyeccion_10 = {"nombre": 1, "_id": 0}

resultado_10 = collection.find(filtro, proyeccion_10)

print("Estudiantes inscritos en SQL o MongoDB:")
for estudiante in resultado_10:
    print(f"- {estudiante['nombre']}")
```
